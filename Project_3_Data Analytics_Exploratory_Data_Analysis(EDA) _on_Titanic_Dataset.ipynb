```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
titanic = sns.load_dataset('titanic')
print("✅ Titanic dataset loaded successfully!\n")
print(titanic.head())
```

```python
print("\n--- Dataset Info ---")
print(titanic.info())

print("\n--- Summary Statistics ---")
print(titanic.describe(include='all'))
```

```python
print("\n--- Missing Values ---")
print(titanic.isnull().sum())
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(titanic.isnull(), cbar=False, cmap='mako')
plt.title("Missing Values in Titanic Dataset")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.histplot(titanic['age'].dropna(), bins=30, kde=True, color='skyblue')
plt.title("Age Distribution of Passengers")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='pclass', data=titanic, palette='Set2')
plt.title("Passenger Count by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='survived', data=titanic, palette='coolwarm')
plt.title("Survival Count (0 = Died, 1 = Survived)")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='sex', hue='survived', data=titanic, palette='Set1')
plt.title("Survival by Gender")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='pclass', hue='survived', data=titanic, palette='Set3')
plt.title("Survival by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.violinplot(x='survived', y='age', data=titanic, palette='muted')
plt.title("Age Distribution by Survival")
plt.xlabel("Survived (0=No, 1=Yes)")
plt.ylabel("Age")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Titanic Dataset")
plt.show()
```

```python
print("\n✅ Key Insights from EDA:")
print("1️⃣ Female passengers had a much higher survival rate than males.")
print("2️⃣ Passengers in 1st class were more likely to survive than those in 2nd or 3rd class.")
print("3️⃣ Younger passengers tended to survive slightly more often.")
print("4️⃣ Fare was positively correlated with survival (wealthier passengers survived more).")
```
